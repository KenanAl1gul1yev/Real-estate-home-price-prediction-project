```python
from bs4 import BeautifulSoup
import requests

import pandas as pd
import sqlalchemy

Server = "localhost"
Database = "Homeprices_Baku"
Driver = "ODBC Driver 17 for SQL Server" # in order to get this info, click Win + S and write 'odbc', then open drivers menu and scroll down
engine = sqlalchemy.create_engine(f'mssql://@{Server}/{Database}?driver={Driver}')
```

```python
baseurl = 'https://bina.az'
headers = {
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36',
}

for p in range(1, 420):
    r = requests.get(f'https://bina.az/alqi-satqi/heyet-evleri?page={p}', headers = headers)
    soup = BeautifulSoup(r.content, 'lxml')
    productlist = soup.find_all('div', class_ = 'card_params')
    
    products = {
    'Location' : [],
    'Rooms' : [],
    'Area' : [],
    'Price' : []
    }
    
    for item in productlist:
        price_text = item.find('span', class_ = 'price-val').get_text(strip = True)
        loc_text = item.find('div', class_ = 'location').get_text(strip = True)
        
        products['Price'].append(int("".join(price_text.split())))
        products['Location'].append(loc_text)
             
        ul_tag = item.find('ul', class_ = 'name')
        li_elements = ul_tag.find_all('li')
        if len(li_elements) == 0:
            products['Rooms'].append(0)
            products['Area'].append(0)
        elif len(li_elements) == 1:
            if 'otaqli' in li_elements[0]:
                products['Rooms'].append(int("".join(li_elements[0].get_text(strip = True).split()[0])))
                products['Area'].append(0)
            else:
                products['Rooms'].append(0)
                products['Area'].append(float("".join(li_elements[0].get_text(strip = True).split()[0])))
        else:
            products['Rooms'].append(int("".join(li_elements[0].get_text(strip = True).split()[0])))
            products['Area'].append(float("".join(li_elements[1].get_text(strip = True).split()[0])))
    df = pd.DataFrame(products)
    df.to_sql(
        name = 'Houses',
        con = engine,
        index = False,
        if_exists = "append",
        dtype={'Location': sqlalchemy.types.Unicode}
    )
    print("ADDED PAGE_", p)
```
